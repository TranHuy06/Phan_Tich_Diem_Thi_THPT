VISUALIZATION

OverView Analysis

```python
# Nhóm dữ liệu theo năm
years = df_all_data['Year'].unique()  # Lấy danh sách các năm
df_all_data_subject = df_all_data.drop(columns='SBD')

# Xác định số lượng hàng và cột (mỗi hàng là một năm, mỗi cột là một môn học)
subjects = [col for col in df_all_data_subject.columns if col != 'Year'] 
nrows = len(years)  # Số hàng tương ứng với số Year
ncols = len(subjects)  # Số cột là số môn học

# Tạo figure và axes cho biểu đồ
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20, 15))  # Tạo lưới

# Lặp qua các Year và vẽ biểu đồ cho từng môn học theo từng Year
for row, year in enumerate(years):
    # Lọc dữ liệu cho từng Year
    df_year = df_all_data_subject[df_all_data_subject['Year'] == year]
    
    for col, subject in enumerate(subjects):
        # Vẽ biểu đồ histogram cho môn học hiện tại trong Year hiện tại
        axes[row, col].hist(df_year[subject], bins='auto', rwidth=1)
        
        # Đặt tiêu đề cho từng biểu đồ
        axes[row, col].set_title(f'{subject} - {year}', fontsize=8)
        
        # Định dạng trục x và y cho biểu đồ
        axes[row, col].tick_params(axis='x', labelsize=6)
        axes[row, col].tick_params(axis='y', labelsize=6)

# Căn chỉnh bố cục
plt.tight_layout()
plt.show()
```
